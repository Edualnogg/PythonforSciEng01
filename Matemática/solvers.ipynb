USO CORRETO DOS SOLVERS:

linsolve((A,B),[x,y,z]) A = matriz /  B = vetor resultado
                 |-> lista das variáveis que desejo saber

```python
#EX:
import sympy as sp

AA = sp.Matrix([[3,2,4],
                [1,1,2],
                [4,3,-2]])
AA
```

```python
y1,y2,y3 = sp.symbols(['y1','y2','y3'])

Y = sp.Matrix([[y1,y2,y3]]).T
Y
```

```python
RR = sp.Matrix([[1,2,3]]).T
RR
```

```python
#ao invés de fazer igual ao solve normal | AA*Y - RR |
#fazer:

sp.linsolve((AA,RR),[y1,y2,y3]) #retornará os valores das variáveis (y1,y2,y3) --->
```

solveset( eq, args ) ou solve

Método nonlinsolve(eq,(arg))

```python
sist = AA*Y - RR
sp.nonlinsolve(sist,(y1,y2,y3))
```

```python
import sympy as sp
sp.init_printing()


x = sp.symbols('x')

f_x = x**3 + 5*x**2 + x

sp.solve(f_x,x) #usando o método acima
```

                                Resolvendo um sisteminha linear

```python
A = sp.Matrix([[3,2,4],[1,1,2],[4,3,-2]])

x1,x2,x3 = sp.symbols(['x1','x2','x3'])

X = sp.Matrix([x1,x2,x3])
X #assim retorna em forma de vetor
```

```python
X = sp.Matrix([[x1,x2,x3]]).T
X
```

```python
R = sp.Matrix([[1,2,3]]).T
R
```

```python
# AX = R
#porém no solve:
#AX - R = 0   ---> o sistema já entende que a equação está igualada a zero, portanto:

sistema = A*X - R
sistema
```

```python
res = sp.solve(sistema,(x1,x2,x3)) #retorna os valores das variáveis da tupla
res
```

```python
res[x1] #dá pra chamar as variáveis específicas
```

```python
#resolvendo um sisteminha não linear:

A_n1 = sp.Matrix([[x1 + x2 - 3],[x1**2 + x2**2 - 9]])
A_n1
```

O sistema acima se resolve:

x1 + x2 = 3
x1**2 + x2**2 = 9

portanto (x1 = 0 e x2 = 3) OU ao contrário

```python
res_n1 = sp.solve(A_n1,(x1,x2)) #pedindo o resultado de x1 e x2
res_n1
```

```python
res_n1[0] #como o resultado sai em tuplas dentro de uma lista, posso solicitar um item específico
```

dsolve( eq, args ), Function('fx') e Derivative( fx,args )

```python
y_x = sp.Function('y_x')

d2y_x = sp.Derivative(y_x(x),x,x) #derivada de segunda ordem

#y_x(x)  ---> y de x em função de x

#o x,x significa derivada sucessiva, ou seja, a derivada da derivada, ou de segunda ordem

d2y_x 
```

```python
eq = d2y_x - 5
sp.dsolve(eq)
```
